```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
malaria_deaths = pd.read_csv("data/malaria_deaths.csv")
malaria_deaths_age = pd.read_csv("data/malaria_deaths_age.csv")
malaria_inc = pd.read_csv("data/malaria_inc.csv")
```

```python
malaria_deaths.rename(columns={"Deaths - Malaria - Sex: Both - Age: Age-standardized (Rate) (per 100,000 people)": "Rate"}, inplace=True)
```

```python
top_5_codes = malaria_deaths.groupby("Code", as_index=False)["Rate"].mean().sort_values(by="Rate", ascending=False).reset_index(drop=True).loc[:4, "Code"].to_list()
```

```python
for code in top_5_codes:
    tmp = malaria_deaths.loc[malaria_deaths["Code"]==code, :].reset_index(drop=True)
    plt.plot(tmp["Year"], tmp["Rate"], label=tmp.loc[0, "Entity"])

plt.legend(loc="right", bbox_to_anchor=(1.4, 0.5))
plt.xlabel("Year")
plt.ylabel("Death rate per 100,000 people")
plt.title("Age-standardized Malaria death rate (per 100,000 people)")
plt.show()
```

```python
age_groups = ["Under 5", "5-14", "15-49", "50-69", "70 or older"]

for age in age_groups:
    tmp = malaria_deaths_age.loc[malaria_deaths_age["age_group"]==age, :].groupby("year", as_index=False)["deaths"].sum()
    plt.plot(tmp["year"], tmp["deaths"], label=age)

plt.legend(loc="right", bbox_to_anchor=(1.3, 0.5))
plt.yscale("log")
plt.xlabel("Year")
plt.ylabel("Total number of deaths")
plt.title("Malaria death by age (global)")
```
